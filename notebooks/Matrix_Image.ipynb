```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Lista de nombres de archivos de imagen
image_files = [
    'RNN_plot_0.png',
    'RNN_plot_1.png',
    'RNN_plot_2.png',
    'RNN_plot_3.png',
    'RNN_plot_4.png',
    'RNN.png'
]

# Etiquetas para las imgenes
labels = ['a)', 'b)', 'c)', 'd)', 'e)', 'f)']

# Crear la figura con un tamao mayor
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Iterar sobre las imgenes y los ejes
for i, (img_file, label) in enumerate(zip(image_files, labels)):
    # Leer la imagen
    img = mpimg.imread(img_file)
    # Calcular la posicin en la matriz
    ax = axes[i // 2, i % 2]
    # Mostrar la imagen
    ax.imshow(img)
    # Quitar los ejes
    ax.axis('off')
    # Aadir la etiqueta
    ax.text(10, 10, label, fontsize=20, color='red', weight='bold', backgroundcolor='white')

# Ajustar el espacio entre subplots
plt.subplots_adjust(wspace=0, hspace=0)

# Guardar la figura resultante
plt.savefig('combined_image_RNN.png', bbox_inches='tight', dpi = 2500)

# Mostrar la figura
plt.show()
```
